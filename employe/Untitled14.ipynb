```python
import pandas as pd

# Sample Employee Dataset
data = {
    "EmployeeID": [201, 202, 203, 204, 205, 206],
    "Name": ["John", "Emma", "Michael", "Sophia", "David", "Olivia"],
    "Department": ["Sales", "HR", "Sales", "IT", "HR", "IT"],
    "Tasks_Completed": [50, 40, 60, 30, 45, 35],
    "Hours_Worked": [160, 150, 170, 140, 155, 145]
}

df = pd.DataFrame(data)
df["Efficiency"] = df["Tasks_Completed"] / df["Hours_Worked"]
df
```

```python
# Highest efficiency employee
top_employee = df.loc[df["Efficiency"].idxmax()]
print("üèÜ Top Employee by Efficiency:\n", top_employee)

# Department-wise efficiency
dept_efficiency = df.groupby("Department")["Efficiency"].mean()
print("\nüìä Department-wise Average Efficiency:\n", dept_efficiency)
```

```python
import matplotlib.pyplot as plt

# Department-wise efficiency bar chart
plt.figure(figsize=(6,4))
dept_efficiency.plot(kind="bar", color="teal")
plt.title("Average Efficiency by Department")
plt.xlabel("Department")
plt.ylabel("Efficiency")
plt.savefig("dept_efficiency.png", dpi=300)
plt.show()

# Employee efficiency scatter plot
plt.figure(figsize=(6,4))
plt.scatter(df["Name"], df["Efficiency"], color="red")
plt.title("Employee Efficiency")
plt.xlabel("Employee")
plt.ylabel("Efficiency")
plt.savefig("employee_efficiency.png", dpi=300)
plt.show()
```
