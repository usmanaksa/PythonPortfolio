```python
import pandas as pd
import numpy as np

# Sample Customer Dataset
data = {
    "CustomerID": [101, 102, 103, 104, 105, 106],
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank"],
    "Age": [25, 30, 35, 28, 40, 22],
    "Country": ["USA", "UK", "USA", "Canada", "UK", "USA"],
    "Purchase_Amount": [250, 300, 150, 400, 500, 100],
    "Purchase_Count": [2, 3, 1, 4, 5, 1]
}

df = pd.DataFrame(data)
df
```

```python
# Total revenue from all customers
total_revenue = df["Purchase_Amount"].sum()
print("üí∞ Total Revenue:", total_revenue)

# Average purchase amount
avg_purchase = df["Purchase_Amount"].mean()
print("üìä Average Purchase Amount:", avg_purchase)

# Top customer by purchase amount
top_customer = df.loc[df["Purchase_Amount"].idxmax()]
print("\nüèÜ Top Customer:\n", top_customer)

# Country-wise revenue
country_revenue = df.groupby("Country")["Purchase_Amount"].sum()
print("\nüåç Revenue by Country:\n", country_revenue)
```

```python
import matplotlib.pyplot as plt

# Country-wise revenue bar chart
plt.figure(figsize=(6,4))
country_revenue.plot(kind="bar", color="orange")
plt.title("Revenue by Country")
plt.xlabel("Country")
plt.ylabel("Revenue ($)")
plt.savefig("country_revenue.png", dpi=300)
plt.show()

# Age vs Purchase Amount scatter plot
plt.figure(figsize=(6,4))
plt.scatter(df["Age"], df["Purchase_Amount"], color="purple")
plt.title("Age vs Purchase Amount")
plt.xlabel("Age")
plt.ylabel("Purchase Amount ($)")
plt.savefig("age_vs_purchase.png", dpi=300)
plt.show()
```
